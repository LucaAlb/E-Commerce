```python
import pandas as pd
```

Pulizia per datamart "Marketing" (CSV "Events" e "Utenti")

```python
utenti= pd.read_csv("../csv/csv_origine_dati/utenti.csv", delimiter=";")
utenti= pd.DataFrame(utenti)
utenti=utenti.rename(columns={"age":"eta", "gender":"genere", "state":"stato", "city":"citta", "country":"paese", "latitude":"latituidne", "longitude":"longitudine", "created_at":"registrato_il"})
utenti.head()
```

```python
minorenni = pd.read_csv("../csv/csv_origine_dati/eventi/minorenni.csv", delimiter = ',') 
minorenni_ = pd.DataFrame(minorenni)
minorenni_=minorenni_.rename(columns={"user_id":"utente_id", "created_at":"accesso alle", "city":"citta", "state":"stato", "traffic_source":"sorgente", "event_type":"evento", "age":"eta"})
minorenni_=minorenni_.drop(columns={"sequence_number", "session_id", "ip_address", "postal_code", "uri"})
minorenni_.head()
```

```python
adulti = pd.read_csv("../csv/csv_origine_dati/eventi/adulti.csv", delimiter = ',')
adulti_ = pd.DataFrame(adulti)
adulti_=adulti_.rename(columns={"user_id":"utente_id", "created_at":"accesso alle", "city":"citta", "state":"stato", "traffic_source":"sorgente", "event_type":"evento", "age":"eta"})
adulti_=adulti_.drop(columns={"sequence_number", "session_id", "ip_address", "postal_code", "uri"})
adulti_.head()
```

```python
senior = pd.read_csv("../csv/csv_origine_dati/eventi/senior.csv", delimiter = ';') 
senior_ =pd.DataFrame(senior) 
senior_=senior_.rename(columns={"user_id":"utente_id", "created_at":"accesso alle", "city":"citta", "state":"stato", "traffic_source":"sorgente", "event_type":"evento", "age":"eta"})
senior_=senior_.drop(columns={"sequence_number", "session_id", "ip_address", "postal_code", "uri"})
senior_.head()
```

```python
senior_plus = pd.read_csv("../csv/csv_origine_dati/eventi/senior_plus.csv", delimiter =';') 
senior_plus_ =pd.DataFrame(senior_plus) 
senior_plus_=senior_plus_.rename(columns={"user_id":"utente_id", "created_at":"accesso alle", "city":"citta", "state":"stato", "traffic_source":"sorgente", "event_type":"evento", "age":"eta"})
senior_plus_=senior_plus_.drop(columns={"sequence_number", "session_id", "ip_address", "postal_code", "uri"})
senior_plus_.head()
```

```python
eventi = pd.concat([minorenni_,adulti_,senior_, senior_plus_])
eventi["evento"].value_counts()
```

```python
eventi_tot= pd.merge(eventi,utenti, left_on="utente_id", right_on="id")
eventi_tot=eventi_tot.drop(columns={"id_x", "id_y", "eta_x", "stato_y", "citta_y"})
eventi_tot=eventi_tot.rename(columns={"citta_x":"citta", "stato_x":"stato", "eta_y":"eta", "accesso alle":"accesso_alle", "latituidne":"latitudine"})
eventi_tot=eventi_tot[["utente_id", "registrato_il", "eta", "genere", "citta", "stato", "paese", "accesso_alle", "evento", "browser", "sorgente", "latitudine", "longitudine"]]
# eventi_tot.to_csv("../csv/csv_analisi/eventi_tot.csv", index=False)
```

Pulizia per datamart "Vendite" (CSV "Ordini", "Ordini_prodotti" e "Utenti")

```python
ordini= pd.read_csv("../csv/csv_origine_dati/ordini.csv", delimiter=";")
ordini_prodotti= pd.read_csv("../csv/csv_origine_dati/ordini_prodotti.csv", delimiter=";")

```

```python
ordini_tot=pd.merge(ordini_prodotti, ordini, left_on="user_id", right_on="user_id")
ordini_tot.head()
```

```python
vendite= pd.merge(ordini_tot, utenti, left_on="user_id", right_on="id")
vendite.info()
```

```python
vendite_= vendite.rename(columns={"order_id_x":"ordini_id", "user_id": "utenti_id", "product_id": "prodotto_id","inventory_item_id":"inv_prod_id","status":"stato_ordine", "created_at_x":"ordinato_il", "shipped_at":"spedito_il","delivered_at": "consegnato_il","returned_at":"reso_il","sale_price":"ptc_sconto","num_of_item": "colli","age":"eta","gender":"genere","state":"stato","city":"citta","country":"paese","latitude":"latitudine", "longitude":"longitudine","created_at_y":"registrato_il"})
vendite_= vendite_.drop(columns={"id_x","id_y","order_id_y"})

vendite_.info()
```

```python
vendite_= vendite_[["utenti_id","eta","genere","registrato_il","stato","paese", "citta", "latitudine","longitudine","ordini_id","prodotto_id","inv_prod_id", "ptc_sconto", "colli","stato_ordine","ordinato_il","spedito_il", "consegnato_il", "reso_il"]]
# vendite.to_csv("../csv/csv_analisi/vendite.csv",index = False)
```

Pulizia per datamart "Magazzino" (CSV "Inventario" e "Centri_distribuzione")

```python
inventario_= pd.read_csv("../csv/csv_origine_dati/inventario_.csv", delimiter=";")
inventario_=inventario_.rename(columns={"product_id":"prodotto_id", "created_at":"caricato il", "sold_at":"venduto_il", "cost":"costo", "product_category":"categoria_prodotto", "product_name":"nome_prodotto", "product_retail_price":"prezzo", "product_department":"dipart_prod", "product_distribution_center_id":"id_cen_distr"})
inventario_.head()
```

```python
centri_distribuzione = pd.read_csv("../csv/csv_origine_dati/centri_distribuzione.csv", delimiter =';') 
centri_distribuzione =pd.DataFrame(centri_distribuzione) 
```

```python
magazzino=pd.merge(inventario_, centri_distribuzione, left_on= "id_cen_distr", right_on= "id")
magazzino=magazzino.drop(columns={"id_x", "id_cen_distr"})
magazzino=magazzino.rename(columns={"id_y":"id_centr_distr"})
# magazzino.to_csv("../csv/csv_analisi/magazzino.csv", index=False)
```
